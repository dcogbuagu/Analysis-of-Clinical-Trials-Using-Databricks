File Extraction Codes

Setting up variables for reusability

```
fileroot1 = 'clinicaltrial_2021'

fileroot2 = 'pharma'

year = '2021'
```

Viewing the files paths to confirm the location of the files

```
# For clinicaltrial

dbutils.fs.ls("dbfs:/FileStore/tables/"+fileroot1+".csv")
```

```
#For pharma

dbutils.fs.ls("dbfs:/FileStore/tables/"+fileroot2+".csv")
```

Copying the files from DBFS to the tmp folder in the local filesystem

```
dbutils.fs.cp("/FileStore/tables/" +fileroot1+ ".zip", "file:/tmp/")

dbutils.fs.cp("/FileStore/tables/" + fileroot2 + ".zip", "file:/tmp/")
```

Checking the tmp directory to see that the files were successfully copied

```
%sh
ls /tmp/
```

Making the varaibles accessible to the shell command line (the local filesystem)

```
import os
os.environ['fileroot1'] = fileroot1

os.environ['fileroot2'] = fileroot2
```

Unzipping the files using the defined variables

```
%sh
unzip -d /tmp /tmp/$fileroot1.zip
```

```
%sh
unzip -d /tmp /tmp/$fileroot2.zip
```

Moving the unzipped files from tmp directory to the DBFS

```
dbutils.fs.mv("file:/tmp/" + fileroot1 + ".csv", "/FileStore/tables/", True)
```

```
dbutils.fs.mv("file:/tmp/" + fileroot2 + ".csv", "/FileStore/tables/", True)
```

Checking the DBFS directory to confirm that the file was successfully moved

```
dbutils.fs.ls("dbfs:/FileStore/tables")
```

To view the content of the files using dbutils.fs.head

```
dbutils.fs.head("dbfs:/FileStore/tables/"+fileroot1+".csv")
```

```
dbutils.fs.head("dbfs:/FileStore/tables/"+fileroot2+".csv")
```
